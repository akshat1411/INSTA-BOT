# Executing the script to scroll down the followers popup until we have atleast 500 followers on the page. Also making a list to store these first 500 followers

```python
follower_Body  = driver.find_element_by_xpath("//div[@class='isgrP']")
scroll = 0
follower_List  = (driver.find_elements_by_xpath("//div[@class='isgrP']//li"))
followers = []
while len(follower_List) < 500:
    driver.execute_script('arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].offsetHeight;', follower_Body)
    time.sleep(0.1)
    follower_List  = driver.find_elements_by_xpath('//div[@class="isgrP"]//li//a[contains(@class,"notranslate")]')
len(follower_List)
```

```python
count = 0
for i in follower_List:
    if count < 500:
        followers.append(i.text)
    else:
        break
    count = count + 1    
followers    
```
